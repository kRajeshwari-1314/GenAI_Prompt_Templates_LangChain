## Task 4: ChatPromptTemplate System

### Description:
This task demonstrates chat-based prompting using system and user roles.

### Approach:
- Used ChatPromptTemplate
- Defined system roles (teacher, interviewer, motivator)

### Output:
Different responses based on assigned roles.

```python
 # System role → defines behavior | User role → dynamic input
# Supported roles: teacher, interviewer, motivator

# Define system behavior descriptions per role (no hardcoding in function)
role_descriptions = {
    "teacher":     "You are an expert teacher. Explain concepts clearly with examples and step-by-step guidance.",
    "interviewer": "You are a professional interviewer. Ask insightful, thought-provoking questions about the topic.",
    "motivator":   "You are an enthusiastic motivator. Inspire the user to learn and explore the topic with energy."
}

def build_chat_prompt(role, topic):
    """
    Build a ChatPromptTemplate dynamically based on role and topic.
    Returns the formatted message list.
    """
    if role not in role_descriptions:
        raise ValueError("Role must be one of: teacher, interviewer, motivator")

    # System message template – injects role behavior dynamically
    system_template = SystemMessagePromptTemplate.from_template(
        "{role_description}"
    )

    # Human/User message template – dynamic topic input
    human_template = HumanMessagePromptTemplate.from_template(
        "Please help me understand {topic}."
    )

    # Combine into a ChatPromptTemplate
    chat_prompt = ChatPromptTemplate.from_messages([system_template, human_template])

    # Format the prompt with actual values
    formatted = chat_prompt.format_messages(
        role_description=role_descriptions[role],
        topic=topic
    )
    return formatted


test_roles = [
    {"role": "teacher",     "topic": "Neural Networks"},
    {"role": "interviewer", "topic": "Deep Learning"},
    {"role": "motivator",   "topic": "Artificial Intelligence"},
]

for entry in test_roles:
    messages = build_chat_prompt(**entry)
    print(f"\n[Role: {entry['role'].upper()} | Topic: {entry['topic']}]")
    for msg in messages:
        print(f"  {msg.type.upper()}: {msg.content}")
```

## Task 5: Input Validation Layer

### Description:
This task ensures that user inputs are valid before generating prompts.

### Approach:
- Created validation function
- Checked allowed values for audience and tone
- Assigned default values if invalid

### Output:
Prevents incorrect inputs and ensures consistency.

```python
# Function Definition
def validate_inputs(audience, tone, strict=False):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Validate Audience
    if audience not in valid_audience:
        if strict:
            raise ValueError(f"Invalid audience '{audience}'. Must be one of: {valid_audience}")
        else:
            print(f"  [WARNING] Invalid audience '{audience}'. Using default: 'beginner'")
            audience = "beginner"

    # Validate Tone
    if tone not in valid_tone:
        if strict:
            raise ValueError(f"Invalid tone '{tone}'. Must be one of: {valid_tone}")
        else:
            print(f"  [WARNING] Invalid tone '{tone}'. Using default: 'casual'")
            tone = "casual"

    return audience, tone


# -----------------------------
# Test 1 - Valid Inputs
# -----------------------------
print("[Test 1 - Valid inputs]")

audience, tone = validate_inputs("beginner", "fun")
print(f"  Validated: audience={audience}, tone={tone}")


# -----------------------------
# Test 2 - Invalid Inputs (Default Handling)
# -----------------------------
print("\n[Test 2 - Invalid inputs with defaults]")

audience, tone = validate_inputs("child", "dramatic")
print(f"  Validated: audience={audience}, tone={tone}")


# -----------------------------
# Test 3 - Invalid Inputs (Strict Mode - Error)
# -----------------------------
print("\n[Test 3 - Invalid inputs raising ValueError]")

try:
    audience, tone = validate_inputs("child", "dramatic", strict=True)
    print(f"  Validated: audience={audience}, tone={tone}")
except ValueError as error:
    print(f"  Caught ValueError: {error}")
```

## Task 6: Prompt Generator App

### Description:
This task integrates all components to generate prompts dynamically.

### Approach:
- Combined validation and templates
- Created a function to generate prompts based on style

### Output:
User-friendly prompt generation system.

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} step by step for {audience} in a {tone} tone"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask 3 interview questions about {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )

    else:
        return "Invalid style"

    return template.format(topic=topic, audience=audience, tone=tone)

# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Final Pipeline Flow

User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output

This pipeline demonstrates how structured prompt engineering works in real-world GenAI applications.

```python
def full_pipeline(topic, audience, tone, style):

    print("=" * 50)
    print("MINI PROMPT ENGINE – Full Pipeline")
    print("=" * 50)

    print(f"[Step 1] User Input   → topic={topic}, audience={audience}, tone={tone}, style={style}")

    # Step 2: Validation (FIXED HERE)
    validated_audience, validated_tone = validate_inputs(audience, tone, strict=False)
    print(f"[Step 2] Validation   → audience={validated_audience}, tone={validated_tone}")

    # Step 3: Template selection
    final_prompt = generate_prompt(topic, validated_audience, validated_tone, style)
    print(f"[Step 3] Template     → Style '{style}' template selected")

    # Step 4: Output
    print(f"[Step 4] Final Output → {final_prompt}")

    print("=" * 50)

    return final_prompt


# Run the pipeline
full_pipeline(
    topic="Artificial Intelligence",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

```
